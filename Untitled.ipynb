# Importing Libraries

```python
import pandas as pd
import numpy as np
%matplotlib inline
from matplotlib import pyplot as plt
```

# Reading Data File

```python
nf = pd.read_csv('Netflix Shows.csv', encoding='latin-1')
nf.head(10)
```

# Remove Duplicate Rows

```python
df = nf.drop_duplicates()
df
```

# Renaming Column Names

```python
df.rename(columns = {'ratingLevel':'ratingDescription','ratingDescription':'ratingLevel' }, inplace = True)  
df
```

```python
df = df.set_index(np.arange(0,500))
```

# Describe DataFrame

```python
df.describe()
```

# Handling missing Values

## Rating Level, Rating Year, User Rating, User Rating Size

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
imputer.fit(df.iloc[:, 3:7])
df.iloc[:, 3:7] = imputer.transform(df.iloc[:, 3:7])
```

 ## Title, Rating, Rating Description

```python
df['title'].fillna("Not_Filled", inplace = True)
# df['Rating'].fillna("Not_Filled" , inplace = True)
df['ratingDescription'].fillna("Not_Filled", inplace = True)
```

# Data Visualization

## In which year highest movies were released?

```python
print('Year wise distribution of netflix shows?')
noOfMoviesInYear  = df['release year'].value_counts().sort_values(ascending = False)
plt.figure(figsize  = (12,4))
plt.title('Years with the number of Netlfix Shows')
noOfMoviesInYear.plot(kind = 'bar')
```

Conclusion : in 2016 highest shows and movies were released i.e 143

## Analysing rating distribution

```python
plt.figure(figsize  = (8, 8))
ratingCount = df['rating'].value_counts()
ratingCount.plot(title = "Analysing rating distribution", kind = 'pie', autopct = '%.f')
```

 Conclusion : TV-MA is the highest rating given
